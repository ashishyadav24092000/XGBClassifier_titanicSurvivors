```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/home/ashish/projects/TItanicSurvivors(XGBoosting)/titanic/train.csv")
data
```

```python
dummies = pd.get_dummies(data['Sex'])
data = pd.concat([data,dummies],axis=1)
data
```

```python
x = data[['Age','female','male']]  ## independent variables
y = data['Survived']  ##dependent variable

seed = 42
test_size = 0.3
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=test_size,random_state=seed)
```

```python
## This is our training data(independent variable)
x_train
```

```python
## Survivor result for the training data
y_train
```

```python
## the test data(independent variables)
x_test
```

```python
## Actual Result of test data
y_test
```

```python
## max_depth sets the maximum depth of the decision trees
model = XGBClassifier(subsample=0.7,max_depth=4,use_label_encoder=False)
model.fit(x_train,y_train)
print(model)
```

```python
predicted_y = model.predict(x_test)
accuracy = accuracy_score(y_test,predicted_y)
print("Accuracy is ::",100*accuracy)
```

```python
## Let's have a look at confusion matrix for above XGBClassifier problem
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predicted_y))  ## displaying the confusion matrix
```
